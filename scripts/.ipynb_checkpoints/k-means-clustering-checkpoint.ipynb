# K-means Clustering
Cluster analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Change file name (path) as necessary
# Read the data
df = pd.read_csv("../data/df_cleaned.csv", sep=',')

pd.set_option('display.max_columns', 47)
#df.head()
```

```python
# Check data dypes
df.dtypes
```

```python
# Drop categorical features
categorical = df.select_dtypes(include=['object']).columns
df = df.drop(categorical, axis=1)
#df.head()
```

```python
# Create scaled DataFrame where each variable has mean of 0 and standard dev of 1
scaled_df = StandardScaler().fit_transform(df)
```

### Finding optimal number of clusters

```python
# Initialize K-Means parameters
kmeans_kwargs = {
    "init": "random",   # Initialization technique
    "n_init": 10,       # The number of initializations to perform
    "random_state": 1   # An integer value to make the results of the algorithm reproducible
}

# Create a list to hold SSE (sum of squared errors) values for each k
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(scaled_df)
    sse.append(kmeans.inertia_)
    
# Visualize results
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

### Perform K-Means Clustering with optimal K
Based on the figure above, there is an elbow (bend) at k = 5 clusters, therefore we will use 5 clusters to fit our K-Means model.

```python
# Instantiate the K-Means class, using optimal number of clusters
kmeans = KMeans(n_clusters=5, **kmeans_kwargs)
```

```python
# Fit K-Means algorithm to data
kmeans.fit(scaled_df)
```

```python
# Append cluster assingments to original dataframe
df['cluster'] = kmeans.labels_
#df.head()
```

### Draw histograms for each cluster
How many students are there in each cluster:
* pre-pandemic (immatriculation before 2018)?
* during the pandemic (immatriculation between 2018 and 2020)?
* after the pandemic (immatriculation after 2020)?

```python
df_cluster_0 = df[df['cluster'] == 0]
df_cluster_1 = df[df['cluster'] == 1]
df_cluster_2 = df[df['cluster'] == 2]
df_cluster_3 = df[df['cluster'] == 3]
df_cluster_4 = df[df['cluster'] == 4]
```

```python
# Students per immatriculation year in cluster 0
df_cluster_0['year_immatriculation'].value_counts().plot(kind='bar')
```

```python
# Students per immatriculation year in cluster 1
df_cluster_1['year_immatriculation'].value_counts().plot(kind='bar')
```

```python
# Students per immatriculation year in cluster 2
df_cluster_2['year_immatriculation'].value_counts().plot(kind='bar')
```

```python
# Students per immatriculation year in cluster 3
df_cluster_3['year_immatriculation'].value_counts().plot(kind='bar')
```

```python
# Students per immatriculation year in cluster 4
df_cluster_4['year_immatriculation'].value_counts().plot(kind='bar')
```

As can be seen from the histograms above, in most clusters the largest amount of students was from a specific time period (either pandemic or pre-/post-pandemic). Only a small amount students were immatriculated in other time periods.

This is not surprising, as the methods of teaching and learning differed greatly in these periods.
